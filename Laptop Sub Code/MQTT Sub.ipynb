```python
import paho.mqtt.client as mqtt
import json
import time
```

```python
#works fine with the raspberry pi thing
topic = ["facility/RPi_Sensor/Values", 'BCV/esp8266_1/Sensor/Val']
json_files = ["RaspberryPi_Json_Data", "BCV1_Json_Data"]


def on_connect(client, userdata, flags, rc):
    """ The callback for when the client receives a CONNACK response from the server."""
    print('Connected with result code ' + str(rc))
    for tpc in topic:
        client.subscribe(tpc)

def on_message(client, userdata, msg):
    """The callback for when a PUBLISH message is received from the server."""

    if msg.topic == topic[0]: #raspberry pi data
        json_file = json.loads(msg.payload.decode())
        print(msg.topic + ' ' + str(json_file))
        with open(json_files[0], 'a+') as outfile:
            outfile.write("\n")
            json.dump(json_file, outfile)

    elif msg.topic==topic[1]: #BCV1 data
        json_file = json.loads(msg.payload.decode())
        print(msg.topic + ' ' + str(json_file))
        with open(json_files[1], 'a+') as outfile:
            outfile.write("\n")
            json.dump(json_file, outfile)


def main():
    mqtt_client = mqtt.Client()
    mqtt_client.on_connect = on_connect
    mqtt_client.on_message = on_message

    mqtt_client.connect("192.168.0.123", 1883)
    mqtt_client.loop_forever()

if __name__ == '__main__':
    print('ABPAS Project')
    main()
time.sleep(0.5)
```
